```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dati = pd.read_csv("./mnist_784_csv.csv")
dati.head()
```

```python
pixel_columns = list(dati.columns)[:-1]
digits = dati[pixel_columns].values
digits
```

```python
digits_names = dati["class"].values
digits_names
```

### Esercizio
sorteggiare 25 cifre a caso nel dataset e rappresentarle graficamente in una figura. Indicare nel titolo di ciascuna la cifra rappresentata

```python
cfr = np.random.choice(np.arange(0, len(digits_names)), 25)
cfr
```

```python
choosen_cfr = digits[cfr]
choosen_names = digits_names[cfr]
```

```python
fig, axs = plt.subplots(nrows=5,ncols=5, figsize=(12,12))
for i,ax in enumerate(axs.reshape(-1)):
    ax.imshow(choosen_cfr[i].reshape((28,28)), cmap="Greys")
    ax.set_axis_off()
    ax.set_title(choosen_names[i],fontsize=8)
plt.show()
```

### Esercizio
sorteggiare 10 scritture per ogni cifra

(usare ciclo for e no reshape degli assi)

```python
dizionario = {}
for c in range(0,10):
    dizionario[c] = digits[digits_names==c]
```

```python
for c in range(0,10):
    immaginiselect = dizionario[c]
    num = immaginiselect.shape[0]
    indici = np.random.choice(np.arange(0, num), 10, replace = False)
    immaginiselect = immaginiselect[indici]
```

```python
fig, axs = plt.subplots(nrows=10,ncols=10, figsize=(12,12))
for c in range(0,10):
    immaginiselect = dizionario[c]
    num = immaginiselect.shape[0]
    indici = np.random.choice(np.arange(0, num), 10, replace = False)
    immaginiselect = immaginiselect[indici]
    for col,img in enumerate(immaginiselect):
        axs[c,col].set_axis_off()
        axs[c,col].set_title(c,fontsize=8)
        axs[c,col].imshow(img.reshape((28,28)), cmap="Greys")
plt.show()
```
